## convert Time in Seconds

````python
def seconds(j):
    """
    This Function will transform the time to seconds. 
    Input = Column Data Frame 00:00:==
    Output values in second 7800
    """
    a = [str(i).split(':') for i in j]
    b = [int(i[0])*3600+int(i[1])*60+int(i[2]) for i in a]
    return b
   
````

## Time in Seconds every 5 K

````python
def seconds5K(j):
    """ 
    This function will take the time each 5 kilometers and calculate the pace per kilometter
    Input = DataFrame with time each 5 km in seconds
    Output: DataFrame pace per kilomter.
    
    """
    seconds_to_minute  = 60
    df = j.copy()
    df = df[['Name','5K', '10K', '15K', '20K', '25K', '30K', '35K','40K', 'Official Time']]
    df = df.rename(columns={'Official Time': '42K'})
    
    for i,k in enumerate(df.columns): 
        if i == 0:
            pass
        elif i==1:
            df[5] = ((df[k])/5)
        elif i == 2:
            df[10] = ((df[k]-df['5K'])/5)
        elif i == 3:
            df[15] = ((df[k]-df['10K'])/5)
        elif i == 4:
            df[20] = ((df[k]-df['15K'])/5)
        elif i == 5:
            df[25] = ((df[k]-df['20K'])/5)
        elif i == 6:
            df[30] = ((df[k]-df['25K'])/5)
        elif i == 7:
            df[35] = ((df[k]-df['30K'])/5)
        elif i == 8:
            df[40] = ((df[k]-df['35K'])/5)
        elif i == 9:
            df[42] = ((df[k]-df['40K'])/2)
    df = df.drop(['5K', '10K', '15K', '20K', '25K', '30K', '35K','40K', '42K'], axis=1)
    return df
````

````python
def minutes5KDElta(j):
    """ 
    This function will take the time each 5 kilometers and calculate the pace per kilometter
    Input = DataFrame with time each 5 km in seconds
    Output: DataFrame pace per kilomter.
    
    """
    df = j.copy()
    
    for i,k in enumerate(df.columns): 
        if i == 0:
            pass
        elif i==1:
            df[5] = pd.to_timedelta('00:'+df[5])
        elif i == 2:
            df[10] = pd.to_timedelta('00:'+df[10])
        elif i == 3:
            df[15] = pd.to_timedelta('00:'+df[15])
        elif i == 4:
            df[20] = pd.to_timedelta('00:'+df[20])
        elif i == 5:
            df[25] = pd.to_timedelta('00:'+df[25])
        elif i == 6:
            df[30] = pd.to_timedelta('00:'+df[30])
        elif i == 7:
            df[35] = pd.to_timedelta('00:'+df[35])
        elif i == 8:
            df[40] = pd.to_timedelta('00:'+df[40])
        elif i == 9:
            df[42] = pd.to_timedelta('00:'+df[42])
    return df
````
